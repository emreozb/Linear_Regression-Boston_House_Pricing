```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Exploratory Data Analysis
##### Distplot 
- Let's first plot the distribution of the target variable (MEDV). Using seaborn to create a distplot.

```python
plt.rcParams["patch.force_edgecolor"] = True
plt.figure(figsize=(10,7))
sns.distplot(boston['MEDV'])
```

- We see that the values of MEDV are normally distributed with few outliers. The histogram of the MEDV doesnt show any severe skewness or extreme outliers.

#### Heatmap 
- Creating a correlation matrix to measure the linear relationships between the variables. Then using seaborn to create a heatmap with the correlation matrix.

```python
plt.figure(figsize=(15,10))
heat_boston = sns.heatmap(boston.corr(), cmap = 'coolwarm', annot = True)
bottom, top = heat_boston.get_ylim()
heat_boston.set_ylim(bottom + 0.5, top - 0.5)
```

- The correlation coefficient ranges from -1 to 1. If the values is close to 1, it means that there is a strong positive correlation between the two variables. When it is close to -1, the variables have a strong negative correlation.


- By looking at the correlation matrix we can see that RM has a strong positive correlation with MEDV (0.7) where as LSTAT has a high negative correlation with MEDV(-0.74).

#### Linear Model Plot
- Creating linear model plots (using seaborn's lmplot) of MEDV vs RM, and MEDV vs LSTAT.

```python
sns.lmplot(x = 'RM', y = 'MEDV', data = boston)
```

```python
sns.lmplot(x = 'LSTAT', y = 'MEDV', data = boston)
```

## Training and Testing Data

- Now that we've explored the data a bit, let's go ahead and split the data into training and testing sets.


- Setting a variable X equal to the RM and LSTAT and a variable y equal to the  MEDV.

```python
y = boston['MEDV']
```

```python
X = boston[['RM', 'LSTAT']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 5)
```

### Training the Model

Now its time to train our model on our training data!

- Importing LinearRegression from sklearn.linear_model

```python
from sklearn.linear_model import LinearRegression
```

**Creating an instance of a LinearRegression() model named lm.**

```python
lm = LinearRegression()
```

- Training/fitting lm on the training data.

```python
lm.fit(X_train, y_train)
```

- Printing out the coefficients of the model

```python
print('Coefficient: \n', lm.coef_)
```

### Predicting Test Data
Now that we have fit our model, let's evaluate its performance by predicting off the test values!

- Using lm.predict() to predict off the X_test set of the data.

```python
predictions = lm.predict(X_test)
```

- Creating a scatterplot of the real test values versus the predicted values. 

```python
plt.scatter(y_test, predictions, color = 'red', edgecolor = 'black')
```

### Predicting Train Data
Let's evaluate its performance by predicting off the train values!

- Using lm.predict() to predict off the X_train set of the data.

```python
train_predictions = lm.predict(X_train)
```

- Creating a scatterplot of the real train values versus the predicted values. 

```python
plt.scatter(y_train, train_predictions, color = 'red', edgecolor = 'black')
```

## Evaluating the Model

Let's evaluate our model performance by calculating the residual sum of squares and the explained variance score (R^2).

- Calculating the Mean Absolute Error, Mean Squared Error, and the Root Mean Squared Error.

```python
from sklearn import metrics
from sklearn.metrics import r2_score
```

```python
print("The model performance for testing set:")
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
print('R2 score:',  r2_score(y_test, predictions))
```

```python
print("The model performance for training set:")
print('MAE:', metrics.mean_absolute_error(y_train, train_predictions))
print('MSE:', metrics.mean_squared_error(y_train, train_predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_train, train_predictions)))
print('R2 score:', r2_score(y_train, train_predictions))
```

### Residuals

We have gotten a very good model with a good fit. Let's quickly explore the residuals to make sure everything was okay with our data. 

- Plotting a histogram of the residuals and make sure it looks normally distributed. Using seaborn to create a distplot.

```python
sns.distplot((y_test - predictions), bins = 50)
```
